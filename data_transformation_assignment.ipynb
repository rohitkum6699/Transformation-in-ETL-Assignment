# Question 1: Define Data Transformation in ETL and explain why it is important.

# Answer:-->
Data Transformation is the process of converting raw data into a clean, structured, and usable format during the ETL (Extract, Transform, Load) process.

#Importance:

1. Removes errors and inconsistencies

2. Improves data quality

3. Makes data suitable for analysis

4. Ensures data matches business requirements

5. Example: Converting dates into the same format (DD-MM-YYYY).

# Question 3: Difference between Normalization and Standardization

# Answer:


| Feature | Normalization               | Standardization             |
| ------- | --------------------------- | --------------------------- |
| Purpose | Scales data between 0 and 1 | Makes data mean = 0, SD = 1 |
| Formula | (x − min) / (max − min)     | (x − mean) / std dev        |
| Range   | 0 to 1                      | No fixed range              |
| Usage   | When data has no outliers   | When data has outliers      |

# Question 6 : What is One-Hot Encoding? Give an example with the categories: “Red, Blue, Green”.

# Answer:
One-Hot Encoding converts categorical data into binary (0/1) columns.

Example: Categories = Red, Blue, Green

| Color | Red | Blue | Green |
| ----- | --- | ---- | ----- |
| Red   | 1   | 0    | 0     |
| Blue  | 0   | 1    | 0     |
| Green | 0   | 0    | 1     |


Each category gets its own column.

#Question 7 : Explain the difference between Data Integration and Data Mapping in ETL
# Answer:

| Feature | Data Integration                     | Data Mapping                           |
| ------- | ------------------------------------ | -------------------------------------- |
| Meaning | Combines data from different sources | Matches source fields to target fields |
| Purpose | Creates a unified dataset            | Defines how data moves                 |
| Example | Merging sales + customer data        | Mapping Cust_ID → CustomerID           |


In short:

Integration = Combining data

Mapping = Connecting columns

# Question 8 : Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist

# Answer:

Min-Max Scaling uses minimum and maximum values.

If outliers exist:

1. They change min and max

2. Normal values get squeezed

Z-score uses mean and standard deviation, not min/max.

Advantages:

1. Less affected by outliers

2. Maintains data distribution

More stable for analysis

Example:

   If salary = [20k, 25k, 30k, 500k]

500k is an outlier.
Min-Max gets distorted.
Z-score handles it better.
